```python
import os
from selenium.webdriver import Edge  # Correct import for Edge WebDriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.edge.options import Options
import requests
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.actions.wheel_input import ScrollOrigin
from bs4 import BeautifulSoup
```

```python
product_name = input("Enter location for Institutes : ")
user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36 Edg/128.0.0.0'

#institute in pitampura

# Path to your msedgedriver.exe
edge_driver_path = os.path.join(os.getcwd(), 'msedgedriver.exe')

# Create a Service object for Edge
edge_service = Service(edge_driver_path)

# Set up Edge options (e.g., user-agent)
edge_options = Options()
edge_options.add_argument(f'user-agent={user_agent}')

# Initialize the Edge browser using correct WebDriver
browser = Edge(service=edge_service, options=edge_options)
browser.get('https://www.google.com/maps/search/'+product_name.replace(' ','+'))

# Initialize ActionChains for scrolling
action = ActionChains(browser)

# Find the first element on the page to start scrolling from
first_element = browser.find_element(By.CSS_SELECTOR, "div.aIFcqe")  # or any reliable element on the page

# Set initial conditions for scrolling
scroll_pause_time =3  # Wait time for the content to load
previous_count = 0  # To keep track of the number of elements loaded

while True:
    # Find all <a> elements to check the content count
    a_tags = browser.find_elements(By.CSS_SELECTOR, "div.rWbY0d")
    current_count = len(a_tags)

    # If no new content is loaded, break the loop
    if current_count == previous_count:
        print("No more content loaded.")
        break

    # Update the previous count to the current count
    previous_count = current_count

    # Scroll down using ActionChains
    last_element = a_tags[-1]  # Scroll to the last found element
    scroll_origin = ScrollOrigin.from_element(last_element)
    action.scroll_from_origin(scroll_origin, 0, 2000).perform()

    # Wait for more content to load
    time.sleep(scroll_pause_time)
    #action.scroll_from_origin(scroll_origin, 0, 2000).perform()
    #time.sleep(scroll_pause_time)
    #action.scroll_from_origin(scroll_origin, 0, 2000).perform()

# After scrolling ends, get all <a> elements
a_tags = browser.find_elements(By.TAG_NAME, "a")

url_list=[]
for a in a_tags:
    url_list.append(a.get_attribute("href"))

maps_urls = [url for url in url_list if url.startswith("https://www.google.com/maps/place/")]
len(maps_urls)
```

```python
Title=[]
Address=[]
Booking=[]
Service=[]
Website=[]
Phones=[]
browser = Edge(service=edge_service, options=edge_options)
for url in maps_urls:
    browser.get(url)
    div_element = browser.find_element(By.CSS_SELECTOR, "div.TIHn2 ")
    inner_html = div_element.get_attribute('innerHTML')
    soup = BeautifulSoup(inner_html,'html.parser')
    title=soup.find('h1',attrs={'class':'DUwDvf lfPIob'}).text
    Title.append(title)

    div_element = browser.find_element(By.CSS_SELECTOR, "div.XltNde.tTVLSc")
    inner_html = div_element.get_attribute('innerHTML')
    soup = BeautifulSoup(inner_html,'html.parser')
    button_soup = soup.find('button',attrs={'data-tooltip':'Copy address'})
    if button_soup is None:
        Address.append(" ")
    else:
        address = button_soup.find('div',attrs={'class':'Io6YTe fontBodyMedium kR99db fdkmkc'}).text
        Address.append(address)

    div_element = browser.find_element(By.CSS_SELECTOR, "div.XltNde.tTVLSc")
    inner_html = div_element.get_attribute('innerHTML')
    soup = BeautifulSoup(inner_html,'html.parser')
    button_soup = soup.find('a',attrs={'data-tooltip':'Open booking link'})
    if button_soup is None:
        Booking.append("No")
    else:
        Booking.append("Yes")

    div_element = browser.find_element(By.CSS_SELECTOR, "div.XltNde.tTVLSc")
    inner_html = div_element.get_attribute('innerHTML')
    soup = BeautifulSoup(inner_html,'html.parser')
    button_soup = soup.find('a',attrs={'data-tooltip':'Services'})
    if button_soup is None:
        Service.append("No")
    else:    
        Service.append("Yes")

    div_element = browser.find_element(By.CSS_SELECTOR, "div.XltNde.tTVLSc")
    inner_html = div_element.get_attribute('innerHTML')
    soup = BeautifulSoup(inner_html,'html.parser')
    button_soup = soup.find('a',attrs={'data-tooltip':'Open website'})
    if button_soup is None:
        Website.append(" ")
    else:
        website = button_soup.get('href')
        Website.append(website)

    div_element = browser.find_element(By.CSS_SELECTOR, "div.XltNde.tTVLSc")
    inner_html = div_element.get_attribute('innerHTML')
    soup = BeautifulSoup(inner_html,'html.parser')
    button_soup = soup.find('button',attrs={'data-tooltip':'Copy phone number'})
    if button_soup is None:
        Phones.append(" ")
    else:
        phone = button_soup.find('div',attrs={'class':'Io6YTe fontBodyMedium kR99db fdkmkc'}).text
        Phones.append(phone)
```

```python
import pandas as pd
df = pd.DataFrame({
    'Name':Title,
    'Address':Address,
    'Booking Available':Booking,
    'Service Available':Service,
    'Website Link':Website,
    'Phone Number':Phones
})
```
